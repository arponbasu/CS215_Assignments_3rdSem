We first generate an array of errors, where each error is the difference between the empirical mean obtained by taking the average of N uniformly distributed numbers, and the actual mean (which is 0.5 for an uniform distribution).

```python
import numpy as np
def plot_Unif(N, M = 100):
  retval = []
  for i in range(M):
    retval.append(abs(0.5 - np.mean(np.random.uniform(0.0, 1.0, N), axis = None)))
  return np.array(retval)
```

We now make a box and whisker plot of the errors obtained.

```python
import matplotlib.pyplot as plt
data = []
data.append(plot_Unif(5))
data.append(plot_Unif(10))
data.append(plot_Unif(20))
data.append(plot_Unif(40))
data.append(plot_Unif(60))
data.append(plot_Unif(80))
data.append(plot_Unif(100))
data.append(plot_Unif(500))
data.append(plot_Unif(1000))
data.append(plot_Unif(10000))
data = np.array(data)

bp = plt.figure(figsize=(15,10)).add_axes([0,0,1,1]).boxplot(data)

plt.show()
```

A function similar to the uniform variable case is written for normal variables below.

```python
import numpy as np
def plot_Gaussian(N, M = 100):
  retval = []
  for i in range(M):
    retval.append(abs(np.mean(np.random.normal(0.0, 1.0, N), axis = None)))
  return np.array(retval)
```

We now make a box and whisker plot of the errors obtained.

```python
import matplotlib.pyplot as plt
data = []
data.append(plot_Gaussian(5))
data.append(plot_Gaussian(10))
data.append(plot_Gaussian(20))
data.append(plot_Gaussian(40))
data.append(plot_Gaussian(60))
data.append(plot_Gaussian(80))
data.append(plot_Gaussian(100))
data.append(plot_Gaussian(500))
data.append(plot_Gaussian(1000))
data.append(plot_Gaussian(10000))
data = np.array(data)

bp = plt.figure(figsize=(15,10)).add_axes([0,0,1,1]).boxplot(data)

plt.show()
```
