```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import pandas as pd
from scipy.stats import bernoulli, binom
```

```python
#Poisson Random Variables,
lambda_x = 3
lambda_y = 4
no=1000000
```

```python
X = np.random.poisson( lambda_x, size=(no,1) )
```

```python
Y = np.random.poisson( lambda_y, size=(no,1) )
```

```python
Z = X+Y
```

```python
plt.plot( Z)
```

```python
Z_x, Z_y = np.unique(Z, return_counts=True)
```

```python
plt.plot(Z_x, Z_y)
```

```python
def poisson_pdf(x, lambd):
    var = np.power(lambd, x)*np.exp(-lambd)/factorial(x)
    return var
```

```python
def pdf_plot(function, range_x, param):
    x_axis = np.arange(0, range_x)
    y_axis = function(x_axis, param)
    plt.plot(x_axis, y_axis)
    plt.show()
```

```python
pdf_plot(poisson_pdf, 20, lambda_x+lambda_y)
```

```python
Z_y
```

```python
# plt.plot(Z_x, Z_y/1000000)
pdf_plot(poisson_pdf, 25, lambda_x+lambda_y)
plt.plot(Z_x, Z_y/no)
```

```python
m=0
v = np.zeros(Z_y.shape)
for i in range(0, Z_y.shape[0]):
    v[i]=abs(poisson_pdf(i, lambda_x+lambda_y)- Z_y[i]/1000000)
    print("Error for k =", i, "is - ", v[i] )
    m+=v[i]
print("\nAverage Error- ")
print(m/25)    
```

```python
plt.plot(range(0,Z_y.shape[0]), v)
plt.show()
```
