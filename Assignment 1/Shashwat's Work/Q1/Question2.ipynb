```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import pandas as pd
```

```python
#Poisson Random Variables,
lambda_x = 3
lambda_y = 4
no=1000000
```

```python
X = np.random.poisson( lambda_x, size=(no,1) )
```

```python
Y = np.random.poisson( lambda_y, size=(no,1) )
```

```python
Z = X+Y
```

```python
plt.plot( Z)
```

```python
Z_x, Z_y = np.unique(Z, return_counts=True)
```

```python
plt.plot(Z_x, Z_y)
```

```python
def poisson_pdf(x, lambd):
    var = np.power(lambd, x)*np.exp(-lambd)/factorial(x)
    return var
```

```python
def pdf_plot(function, range_x, param):
    x_axis = np.arange(0, range_x)
    y_axis = function(x_axis, param)
    plt.plot(x_axis, y_axis)
    plt.show()
```

```python
pdf_plot(poisson_pdf, 20, lambda_x+lambda_y)
```
