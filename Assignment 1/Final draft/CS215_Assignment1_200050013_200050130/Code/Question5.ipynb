We first generate an array of errors, where each error is the difference between the empirical mean obtained by taking the average of N uniformly distributed numbers, and the actual mean (which is 0.5 for an uniform distribution).

```python
#All cells combined take around 2 seconds to execute and generate all graphs
import numpy as np
np.random.seed(1)

def plot_Unif(N, M = 100):
  retval = []
  for i in range(M):
    retval.append(abs(0.5 - np.mean(np.random.uniform(0.0, 1.0, N), axis = None)))
  return np.array(retval)
```

We now make a box and whisker plot of the errors obtained. As predicted by the Law of large numbers (which states that as n goes to infinity, mu_hat *strongly converges* towards mu, ie:- events for which mu_hat is not equal to mu have zero probability), we can see the median going to zero, although not monotonically (see N = 60 and N = 80 for example). We also observe a general decreasing trend in the standard deviation.

```python
testValues = [5,10,20,40,60,80,100,500,1000,10000]
data = []
for x in testValues:
  data.append(plot_Unif(x))

data = np.transpose(np.array(data))

import matplotlib.pyplot as plt
labelValues = [0] + testValues
labels = list(map(lambda x : ("N = " + str(x)), labelValues))
labels[0] = '0'
plt.figure(figsize=(15,10)).add_axes([0,0,1,1]).boxplot(data)
plt.xticks(np.arange(len(labelValues)), labels)
plt.xlabel("Size of Random Variable Draw")
plt.ylabel("Box Plot Metrics")
plt.title("Box Plot for the Uniform RV case")
plt.savefig("UniformBoxPlot.png", dpi=100)
plt.show()
```

A function similar to the uniform variable case is written for normal variables below.

```python
def plot_Gaussian(N, M = 100):
  retval = []
  for i in range(M):
    retval.append(abs(np.mean(np.random.normal(0.0, 1.0, N), axis = None)))
  return np.array(retval)
```

We now make a box and whisker plot of the errors obtained.

```python
testValues = [5,10,20,40,60,80,100,500,1000,10000]
data = []
for x in testValues:
  data.append(plot_Gaussian(x))

data = np.transpose(np.array(data))

import matplotlib.pyplot as plt
labelValues = [0] + testValues
labels = list(map(lambda x : ("N = " + str(x)), labelValues))
labels[0] = '0'
plt.figure(figsize=(15,10)).add_axes([0,0,1,1]).boxplot(data)
plt.xticks(np.arange(len(labelValues)), labels)
plt.xlabel("Size of Random Variable Draw")
plt.ylabel("Box Plot Metrics")
plt.title("Box Plot for the Gaussian RV case")
plt.savefig("GaussianBoxPlot.png", dpi=100)
plt.show()
```
